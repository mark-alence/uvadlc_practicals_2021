```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
txt = 'book_EN_democracy_in_the_US'
# txt = "book_EN_grimms_fairy_tails"
accs = np.load(f'accuracies_{txt}.npy')
losses = np.load(f'losses_{txt}.npy')
# losses = losses.reshape(20,-1)
# losses = np.mean(losses, axis=1)
x = np.arange(25)
```

```python
accs
```

```python
plt.plot(x, accs)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title(f"Training accuracy of LSTM over 20 epochs trained on {txt}")
plt.savefig("lstm_acc", bbox_inches='tight')
```

```python
plt.plot(x, losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title(f"Training loss of LSTM over 20 epochs trained on {txt}")
plt.savefig("val_acc",  bbox_inches='tight')
```
